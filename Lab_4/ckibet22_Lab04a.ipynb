```python
cancer_csv = pd.read_csv("https://raw.githubusercontent.com/patrickmlong/Breast-Cancer-Wisconsin-Diagnostic-DataSet/master/data.csv")

#Data shape
print(cancer_csv.shape)

#column headers
print('Column headers:\n', cancer_csv.columns)

#get first 5 items
cancer_csv.head()

#get mean of each column
cancer_csv.mean(axis = 0)

#Cancer_x dataframe with columns with "mean"
cancer_x = cancer_csv[['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean']]

#Normalize cancer_x: subtract min and divide by range
cancer_x = (cancer_x - cancer_x.min())/ (cancer_x.max() - cancer_x.min())

#get diagnosis from cancer_csv as response variables and set to lower case
cancer_y = pd.Series(data = cancer_csv['diagnosis']).str.lower()
```

## Task 2: Create and Visualize a Covariance Matrix

1. Use the [Dataframe](https://pandas.pydata.org/pandas-docs/stable/reference/frame.html) documentation to find out how to compute the pairwise covariance of the columns, store the resulting covariance matrix, and print its shape.
2. Use the [matplotlib](https://matplotlib.org/3.1.1/gallery/images_contours_and_fields/image_annotated_heatmap.html) documentation to visualize the covariance matrix with a heatmap.
    1. Include a title
    2. Include labels for each row and column 
        1. Set the `xticks`, `yticks`, `xticklabels` and `ytickslabels`
        1. rotating the text for the x-axis using `setp`
    3. Use an perceptually uniform sequential [colormap](https://matplotlib.org/tutorials/colors/colormaps.html). 
        1. Set it using the `cmap` parameter for `imshow`
    4. Include the `colorbar`.
    
**Question:** What interpretations can you draw from this covariance matrix?
A covariance matrix can tell us the relationship between two variables when one of the variables changes. We can tell whether the relationship is positive or negative using a heatmap that indicates the depth of color shade based on strength of relationship between variables.

```python
#covariance matrix
cov_mat = cancer_x.cov()

#covariance to numpy
cov_mat = cov_mat.to_numpy()

fig, ax = plt.subplots()
im = ax.imshow(cov_mat, cmap=plt.cm.Blues)

# We want to show all ticks...
ax.set_xticks(np.arange(cov_mat.shape[0]))
ax.set_yticks(np.arange(cov_mat.shape[0]))

# ... and label them with the respective list entries
ax.set_xticklabels(cancer_x.columns)
ax.set_yticklabels(cancer_x.columns)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(cov_mat.shape[0]):
    for j in range(cov_mat.shape[0]):
        text = ax.text(j, i, f'{cov_mat[i, j]: 0.2f}',
                       ha="center", va="center", color="black")
# Create colorbar
cbar = ax.figure.colorbar(im, ax=ax, shrink = 0.70)
cbar.ax.set_ylabel(ylabel = 'Covariance spectrum', rotation=-90, va="bottom")

ax.set_title("Covariance matrix cancer_csv")
fig.tight_layout()
plt.show()
```

## Task 3: Visualizing Principle Component Analysis

### 3a) Principle Component Analysis
1. Use [numpy.linalg.eig](https://docs.scipy.org/doc/numpy/reference/generated/numpy.linalg.eig.html) to compute the eigenvalues and right eigenvectors of the covariance matrix.  
1. Print the eigen_values and the first 2 eigenvectors.
2. Transform the data with the eigen_vectors.
    1. Matrix multiply cancer_x with the first two columns of the eigen_vectors. 
    2. Print the first 5 rows of the transformed data.

**Question:**  Which 2 variables are the first eigenvector most dependent on?  How does that relate to the heatmap?
Radius_mean and perimeter_mean. These two variables exhibit the darkest shade of color in the first column of the heatmap.

```python
#compute eigenvalues and eigenvectors of covariance matrix
eig_vals, eig_vecs = np.linalg.eig(cov_mat)

print(f'Eigen values:\n{eig_vals}')
print(f'First 2 Eigen vectors:\n{eig_vecs[:,0:2]}')

#transform data
trans_data = cancer_x @ eig_vecs[:,0:2]

print(f'Transformed data:\n{trans_data[0:5]}')

print('''
-----------------------------------
Your results should look like this:
-----------------------------------
Eigen values: 
[0.15632 0.05609 0.01875 0.01012 0.00688 0.00348 0.00001 0.00027 0.00121 0.00206]
First 2 Eigen vectors: 
[[-0.36654  0.33267]
 [-0.1315   0.11754]
 [-0.38051  0.30003]
 [-0.32796  0.28817]
 [-0.16268 -0.33468]
 [-0.34395 -0.32018]
 [-0.43907 -0.17375]
 [-0.47861 -0.04037]
 [-0.16356 -0.34204]
 [-0.05141 -0.58234]]
Transformed data:
           0         1
0 -1.692053 -0.746755
1 -1.105772  0.158411
2 -1.479314 -0.159259
3 -1.335070 -1.286536
4 -1.364975 -0.036868
-----------------------------------
''')
```
